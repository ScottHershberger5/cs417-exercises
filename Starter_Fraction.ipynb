```python
class Fraction:
    def __init__(self, top, bottom):
        self.num = top
        self.den = bottom

    # def __str__(self):
    #     return f"{self.num}/{self.den}"

    def __add__(self, other):
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den

        return Fraction(new_num, new_den)



f1 = Fraction(3,4)
f2 = Fraction(4,5)
print(f1)
print(f2)
print(f1 + f2)
```
